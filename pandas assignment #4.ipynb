# Q2.- Make a function for batting profile of any batsman:-                                              [Medium+]


# #Ex- 

V Kohli:

Innings:  

Total Runs: 

No. of MOM(Man of the match):

No. of boundries:

```python
def stat(b):
 mask1=data2['batsman']==b
 data3=data2[mask1]
 Innings=data3['match_id'].value_counts().shape[0]
 Total_Runs=data3['batsman_runs'].sum()
 mask2=data1['player_of_match']==b
 MOM=data1[mask2].shape[0]
 mask3=data3['batsman_runs']==4
 mask4=data3['batsman_runs']==6
 boun=data3[mask3+mask4].shape[0]
 print('Innings:',Innings)

 print('Total Runs',Total_Runs)

 print('No. of MOM(Man of the match):',MOM)

 print('No. of boundries:',boun)
```

```python
stat('RA Jadeja')
```

```python
stat('SR Watson')
```

```python
stat('V Kohli')
```

```python
stat('MS Dhoni')
```

```python
stat('RG Sharma')
```

# Q3.- Write a function to return top-4 batsman at a venue. Take venue as input.

```python
data2.merge(data1,left_on='match_id',right_on='id').info()
```

```python
def venbat(v):
 data3=data2.merge(data1,left_on='match_id',right_on='id')
 data4=data3.groupby(['venue','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index()
 mask11=data4['venue']==v
 return(data4[mask11].head(4))
```

```python
venbat('Eden Gardens')
```
