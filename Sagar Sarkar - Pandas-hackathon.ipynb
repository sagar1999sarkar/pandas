**Problem 1** : Write a python function which will receive 3 input i.e. batch name, start date and end date and returns the leaderboard(in that date window) of that particular batch with the following pieces of info - Name, Marks(till now)

Note: If no date range is provided, show the leaderboard for the entire duration of the course

**Marks -10**

```python
def topper(bn,sd,ed): 
 data1=batch.merge(batch_task,left_on='batch_id',right_on='batch_id')
 data2=data1.merge(evaluation,left_on='task_id',right_on='task_id')
 mask1=data2['batch_name']==bn
 data3=data2[mask1]
 mask2=data3['evaluation_date']>=sd
 mask3=data3['evaluation_date']<=ed
 data4=data3[mask2 & mask3]
 data5=data4.groupby('user_id').sum()  
 data6=data5.merge(users,left_on='user_id',right_on='user_id')
 data7=data6.drop(['bt_id','batch_id_x','bt_id','task_id','e_id','late','winner','batch_id_y','email','dp','college','passing_year','branch','github','medium','portfolio','linkedin','youtube'],axis=1)
 gyan['gs'] = gyan['category'].apply(lambda x: 1 if not pd.isnull(x) else np.nan)
 mask4=gyan['submission_date']>=sd
 mask5=gyan['submission_date']<=ed
 gyan1=gyan[mask4 & mask5]
 gyan2=gyan1.groupby('user_id').sum()   
 gyan2=gyan2.drop(['gyan_id'],axis=1)
 data8=data7.merge(gyan2,right_on='user_id',left_on='user_id')
 myday['ms'] = myday['entry'].apply(lambda x: 1 if not pd.isnull(x) else np.nan)
 mask10=myday['entry_date']>=sd
 mask11=myday['entry_date']<=ed
 my=myday[mask10 & mask11]
 my2=my.groupby('user_id').sum()  
 my2=my2.drop(['myday_id'],axis=1)
 data9=data8.merge(my2,right_on='user_id',left_on='user_id')
 data9['Total_Score']=data9['score']+data9['ms']+data9['gs'] 
 data10=data9.drop(['score','gs','ms'],axis=1)
 data11=data10.sort_values('Total_Score',ascending=False).head(25)
 return(data11)   
```

```python
topper('Machine Learning Kolkata 2','2020-01-09','2020-03-02')
```

```python
def consis(bn,sd,ed):
 data1=batch.merge(batch_task,left_on='batch_id',right_on='batch_id')
 data2=data1.merge(evaluation,left_on='task_id',right_on='task_id')
 mask1=data2['batch_name']==bn
 data3=data2[mask1]
 mask2=data3['evaluation_date']>=sd
 mask3=data3['evaluation_date']<=ed
 data4=data3[mask2 & mask3]
 data5=data4.groupby('user_id').sum()  
 data6=data5.merge(users,left_on='user_id',right_on='user_id')
 data7=data6.drop(['bt_id','batch_id_x','bt_id','task_id','e_id','batch_id_y','email','dp','college','passing_year','branch','github','medium','portfolio','linkedin','youtube'],axis=1)
 gyan['gs'] = gyan['category'].apply(lambda x: 1 if not pd.isnull(x) else np.nan)
 mask4=gyan['submission_date']>=sd
 mask5=gyan['submission_date']<=ed
 gyan1=gyan[mask4 & mask5]
 gyan2=gyan1.groupby('user_id').sum()   
 gyan2=gyan2.drop(['gyan_id'],axis=1)
 data8=data7.merge(gyan2,right_on='user_id',left_on='user_id')
 myday['ms'] = myday['entry'].apply(lambda x: 1 if not pd.isnull(x) else np.nan)
 mask10=myday['entry_date']>=sd
 mask11=myday['entry_date']<=ed
 my=myday[mask10 & mask11]
 my2=my.groupby('user_id').sum()  
 my2=my2.drop(['myday_id'],axis=1)
 data9=data8.merge(my2,right_on='user_id',left_on='user_id')
 data9['Total_Score']=data9['score']+data9['ms']+data9['gs'] 
 data10=data9.drop(['score','gs','ms'],axis=1)
 data11=data10.sort_values('Total_Score',ascending=False)
 data11['consistency']=data11['Total_Score']-2*data11['late']+2*data11['winner']
 df=data11.head(1)
 name=df['fname'] + " "+df['lname']  
 return('the most consistent student in that given daterange and batch',name)
```
